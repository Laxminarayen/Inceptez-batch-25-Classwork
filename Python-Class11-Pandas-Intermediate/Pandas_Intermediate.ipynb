````python
chipo['quantitytenloop'] = None
for i in range(0,chipo.shape[0]):
  #Value Assigning
  chipo['quantitytenloop'][i] = tenpercent(chipo['quantity'][i])
````

````python
#use LOC if you dont want to initialize it first!!
for i in range(0,chipo.shape[0]):
  #Value Assigning
  chipo.loc[i,'quantitytenloop_withloc'] = tenpercent(chipo['quantity'][i])
````

````python
#apply dataframe['colName'].apply(function)
chipo['quantity10'] = chipo['quantity'].apply(tenpercent)
````
