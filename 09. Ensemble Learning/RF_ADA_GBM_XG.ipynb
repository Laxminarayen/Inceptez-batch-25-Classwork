```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "creditcard.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  r"mlg-ulb/creditcardfraud",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", df.head())
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df['Class'].value_counts(normalize=True)
```

```python
#Oversampling
#Undersampling
#Class imbalance Weighting
#Threshold Tuning
```

```python
X = df.drop(columns ="Class")
y = df['Class']
```

```python
train_X, test_X, train_y, test_y = train_test_split(X,y,test_size =0.2,random_state=24)
train_y.value_counts()
```

```python
test_y.value_counts()
```

```python
DecisionTree_imbalanced = DecisionTreeClassifier()
DecisionTree_imbalanced.fit(train_X,train_y)

train_predictions = DecisionTree_imbalanced.predict(train_X)
test_predictions = DecisionTree_imbalanced.predict(test_X)

print("Training Accuracy: ", accuracy_score(train_y,train_predictions))
print("Testing Accuracy: ", accuracy_score(test_y,test_predictions))

print("Training Report: ", classification_report(train_y,train_predictions))
print("Testing Report: ", classification_report(test_y,test_predictions))

print("Confusion Matrix Train:", confusion_matrix(train_y,train_predictions))
print("Confusion Matrix Test:", confusion_matrix(test_y,test_predictions))
```

```python
#HW Assignment 1 - GridSearch Here
DecisionTree_RoughTuned = DecisionTreeClassifier(max_depth=5)
DecisionTree_RoughTuned.fit(train_X,train_y)

train_predictions = DecisionTree_RoughTuned.predict(train_X)
test_predictions = DecisionTree_RoughTuned.predict(test_X)

print("Training Accuracy: ", accuracy_score(train_y,train_predictions))
print("Testing Accuracy: ", accuracy_score(test_y,test_predictions))

print("Training Report: ", classification_report(train_y,train_predictions))
print("Testing Report: ", classification_report(test_y,test_predictions))

print("Confusion Matrix Train:", confusion_matrix(train_y,train_predictions))
print("Confusion Matrix Test:", confusion_matrix(test_y,test_predictions))
```

```python
def Train_fit_report(model,train_X,train_y,test_X,test_y):
  model.fit(train_X,train_y)
  train_predictions = model.predict(train_X)
  test_predictions = model.predict(test_X)

  print("Training Accuracy: ", accuracy_score(train_y,train_predictions))
  print("Testing Accuracy: ", accuracy_score(test_y,test_predictions))

  print("Training Report: ", classification_report(train_y,train_predictions))
  print("Testing Report: ", classification_report(test_y,test_predictions))

  print("Confusion Matrix Train:", confusion_matrix(train_y,train_predictions))
  print("Confusion Matrix Test:", confusion_matrix(test_y,test_predictions))

  return model
```

```python
#HW 2: Do tuning and see if you can get the best variant of RandomForest Algo
%%time
RandomForest = RandomForestClassifier()
RandomForest = Train_fit_report(RandomForest,train_X,train_y,test_X,test_y)
```

## Ada boost Parameters:
1. n_estimators: How many corrective rounds should my model perform?
2. learning_rate: How strongly should each tree correct the previous mistakes? / How much influence  should each tree have when correct past mistakes.
3. base_estimator: How smart should each learner be

```python
ada = AdaBoostClassifier(n_estimators = 100,learning_rate = 0.5)
ada = Train_fit_report(ada,train_X,train_y,test_X,test_y)
```

1. High Learning Rate - Overfitting
2. Too many estimators include the risk overfitting

```python
gbm = GradientBoostingClassifier(n_estimators = 100,learning_rate = 0.1,max_depth=3,random_state = 24)
gbm = Train_fit_report(gbm,train_X,train_y,test_X,test_y)
```

## Observations of GBM results:
1. Slower than Randomforest
2. Careful tuning is very very important

```python
gbm = GradientBoostingClassifier()
gbm = Train_fit_report(gbm,train_X,train_y,test_X,test_y)
```

## PArameters of XGBoost
1. n_estimators - How many trees?
2. learning_rate - How fast do we learn?
3. max_depth - how complex is each tree?
4. subsample - How much data does each tree see?
5. colsample_bytree - How many features per tree?
6. reg_alpha & reg_lamda - How much do we punish complexity
7. scale_pos_wright - Hoe much more importatnt is fraud?
8. min_child_weight - How much data must a branch have to be not pruned?
9. gamma - IS this current split worth it? - forces you to have meaningful improvements only - Minimum benifit requrement.
10. colsample_bylevel - how man feature per tree level?

```python
xgb = XGBClassifier()
xgb = Train_fit_report(xgb,train_X,train_y,test_X,test_y)
```
