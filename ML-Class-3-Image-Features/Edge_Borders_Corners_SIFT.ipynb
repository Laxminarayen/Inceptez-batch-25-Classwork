Edges are points in an image where the intensity (brightness or colour) changes sharply.. Boundaries, textures, countours. etc.,

```python
img = cv2.imread("facade.jpeg",cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img,cmap = 'gray')
```

```python
X = np.array([[-1,0,1],
              [-1,0,1],
              [-1,0,1]], dtype = np.float32) #| edges
```

image.png

```python
Features_X = cv2.filter2D(img, -1, X)
print(Features_X)
```

```python
Features_X
```

```python
plt.imshow(Features_X,cmap = 'gray')
```
