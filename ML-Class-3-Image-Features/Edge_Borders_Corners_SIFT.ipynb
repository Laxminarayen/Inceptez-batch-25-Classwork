```python
# #Canny Edge detection System
# 1. Gaussian Filter - Smoothen image (false Edges)
# 2. Compute Gradients / edges
# 3. Non-Maximum Suppression (NMS)
# 4. Doubling the threshold
```

```python
import cv2
img = cv2.imread("facade.jpeg",cv2.IMREAD_GRAYSCALE)

edges = cv2.Canny(img,100,200)

plt.imshow(edges,cmap = 'gray')
```

```python
corners = cv2.cornerHarris(img,2,3,0.04) #Neighborhood of edges =2, Size of sobel = 3x3, 0.04 - 0.06
```

```python
corners.max()
```

```python
plt.imshow(corners,cmap = 'gray') #Edge-map - Corner response map Like  <P - it is a corner
```

```python
img = cv2.imread("facade.jpeg")

cornerImage = cv2.dilate(corners,None)
img[cornerImage > 0.01 * cornerImage.max()] = [0,0,255] #Strongest detected corner

plt.imshow(img) #BGR
```
