```python
#Question:6 - Day 6 - Swapcase https://www.hackerrank.com/challenges/swap-case/problem?isFullScreen=true
#Algorithmic Thiking..
#1. Loop through the string
#2. i == lower(): -> convert it to upper()
#3. else: convert it to lower()

def swap_case(s):
  swapped = []
  for i in s:
    if i == i.lower():
      swapped.append(i.upper())
    else:
      swapped.append(i.lower())
  return "".join(swapped)
```

```python
def swap_case(s):
  swapped = ""
  for i in s:
    if i == i.lower():
      swapped+=i.upper()
    else:
      swapped+=i.lower()
  return swapped
```

```python
#Question:7 - Day 7: https://www.hackerrank.com/challenges/the-minion-game/problem?isFullScreen=true
#Algorithmic Thiking..


#eg: BANANA
#Stu
#B BA BAN BANA BANAN BANANA

#Algorithm
#1. Loop through each of the string
#2. Check if the character is a vowel / Consonent
#3. the len of the substring from that index as score

def minion_game(string):
  vowels = "AEIOU"
  kevin_score = 0
  stuart_score = 0
  for i in range(0,len(string),1):
    if string[i] in vowels:
      kevin_score += (len(string) -i) #BANANA
    else:
      stuart_score += (len(string) -i)
  if kevin_score > stuart_score:
    print("Kevin", kevin_score)
  elif stuart_score > kevin_score:
    print("Stuart", stuart_score)
  else:
    print("Draw")
```
