```python
def equ(x):
  return 2*x*x + 2
```

```python
def der(x):
  return 4*x
```

```python
num_iteration = 100000
learning_rate = 0.01
cur_x = 1
print("Iteration : 0 ", ": X value : ", cur_x, " => Y value : ", equ(cur_x))
```

```python
y_res = []
x_res = []
change_happened = []
for i in range(num_iteration):
  prev_x = cur_x
  cur_x = cur_x - (der(cur_x) * learning_rate)
  y_val = equ(cur_x)
  change = abs(cur_x - prev_x)
  x_res.append(cur_x)
  y_res.append(y_val)
  change_happened.append(change)
  print("Iteration : ",i, ": X value : ", cur_x, " => Y value : ", y_val)
  print("Dervivative of y : ", der(cur_x))
  if (change <0.0001):
    num_iteration =  i
    break
print("Minimum Value found")
print("X = ", cur_x)
print("Y = ", equ(cur_x))
print("Dervivative of y : ", der(cur_x))
```

```python
import matplotlib. pyplot as plt
plt.plot(x_res)
plt.show()
```

```python
import matplotlib. pyplot as plt
plt.plot(y_res)
plt.show()
```

```python
import matplotlib. pyplot as plt
plt.plot(range(num_iteration+1), change_happened, color= 'orange', marker = 'o')
plt.grid(True)
plt.show()
```
