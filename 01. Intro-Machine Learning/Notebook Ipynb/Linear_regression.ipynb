```python
import numpy as np
```

```python
X = np.array([1,2,3])
y = np.array([2,3,5])
```

```python
X
```

```python
y
```

```python
iteration = 10000
learning_rate = 0.01
m = 1
b = 1
n = len(X)
```

```python
m * X + b

1*1+1, 1*2+1, 1*3+1
```

```python
for i in range(iteration):
  #print("input :", X)
  y_pred = m * X + b
  #print("y  : ", y)
  #print("y_pred : ", y_pred)

  error =  y - y_pred
  #print("Error array : ", error)
  error_sum = abs(np.sum(error))
  #print("Error Sum : ", error_sum)
  SSE = abs(np.sum(error))**2
  #print("SSE : ", SSE)
  MSE = (abs(np.sum(error))**2)/n
  #print("MSE : ", MSE)

  m_gradient = (2/n) * -(np.dot(X, error))
  #print("m_gradient : ", m_gradient)

  b_gradient = (2/n) * -(np.sum(error))
  #print("b_gradient : ", b_gradient)

  m = m - (learning_rate * m_gradient)
  #print("m : ", m)

  b = b - (learning_rate * b_gradient)
  #print("b : ", b)







print("m : ", m)
print("b : ", b)







```
